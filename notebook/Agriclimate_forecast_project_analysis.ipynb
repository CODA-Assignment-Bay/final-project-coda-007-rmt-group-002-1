## Dampak CO2 Emission terhadap Hasil panen di USA

```python
from scipy.stats import pearsonr
import numpy as np

# Hitung korelasi CO2 emission terhadap crop yield per country
correlation_results = []

for country, sub_df in df.groupby("country"):
    if sub_df["average_temperature_c"].nunique() > 1:
        r, p = pearsonr(sub_df["average_temperature_c"], sub_df["crop_yield_mt_per_ha"])
        correlation_results.append({"country": country, "Correlation": r, "P_Value": p})

# Buat dataframe hasil korelasi
corr_df = pd.DataFrame(correlation_results).sort_values(by="Correlation", key=abs, ascending=False)

# Ambil negara dengan korelasi paling tinggi absolut
top_country = corr_df.iloc[0]["country"]
top_country, corr_df.head()
```

Hasil :

---

Negara yang hasil panennya paling dipengaruhi oleh CO emission (dari sisi korelasi tertinggi) adalah USA, dengan korelasi sekitar -0.171 (berarti ada hubungan negatif: CO naik  hasil panen cenderung turun).

---

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Filter data untuk USA
usa_df = df[df["country"] == top_country].copy()

# Siapkan variabel untuk regresi
X = usa_df[["average_temperature_c"]]
y = usa_df["crop_yield_mt_per_ha"]

# Buat dan latih model regresi linier
model = LinearRegression()
model.fit(X, y)

# Prediksi hasil panen dari data CO2
y_pred = model.predict(X)

# Plot regresi
plt.figure(figsize=(8,5))
plt.scatter(X, y, label="Data Aktual", alpha=0.6)
plt.plot(X, y_pred, color="red", label="Regresi Linier")
plt.xlabel("CO Emissions (MT)")
plt.ylabel("Crop Yield (MT/HA)")
plt.title(f"Hubungan CO Emission vs Crop Yield di {top_country}")
plt.legend()
plt.show()

# Koefisien dan intercept
coef = model.coef_[0]
intercept = model.intercept_

coef, intercept
```

Hasil regresi linier untuk USA menunjukkan:

---

CropYield = 0.01867  COEmissions + 2.525

Artinya, setiap kenaikan 1 MT CO rata-rata menurunkan hasil panen sekitar 0.0187 MT/HA.

## Question 1: Faktor yang paling berpengaruh terhadap hasil panen

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Penentuan variabel
features = ['average_temperature_c', 'total_precipitation_mm', 'co2_emissions_mt']
target = 'crop_yield_mt_per_ha'

X = df[features]
y = df[target]

# Korelasi Pearson
correlation = df[features + [target]].corr()[target].drop(target)
print("Korelasi terhadap Crop Yield:")
print(correlation)

# Regresi linier berganda
model = LinearRegression()
model.fit(X, y)

# Koefisien regresi
coef_df = pd.DataFrame({
    "Variable": features,
    "Coefficient": model.coef_
}).sort_values(by="Coefficient", key=abs, ascending=False)

print("\nKoefisien Regresi (nilai absolut terbesar = pengaruh relatif terbesar):")
print(coef_df)
```

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt


# Pilih variabel
features = ['average_temperature_c', 'co2_emissions_mt', 'total_precipitation_mm',]
target = 'crop_yield_mt_per_ha'

X = df[features]
y = df[target]

# Tambahkan konstanta untuk intercept
X = sm.add_constant(X)

# Buat model OLS
model = sm.OLS(y, X).fit()

# Ekstrak hasil
results_df = pd.DataFrame({
    "Variable": model.params.index,
    "Coefficient": model.params.values,
    "P-Value": model.pvalues.values,
    "T-Statistic": model.tvalues.values
})

# Buang konstanta
results_df = results_df[results_df["Variable"] != "const"]

# Ranking berdasarkan pengaruh absolut
results_df["Abs_Coefficient"] = results_df["Coefficient"].abs()
results_df = results_df.sort_values(by="Abs_Coefficient", ascending=False)

# Tampilkan ranking di console
print("\nRanking Pengaruh Variabel terhadap Crop Yield:")
print(results_df)

# Visualisasi
plt.figure(figsize=(8, 5))
plt.barh(results_df["Variable"], results_df["Abs_Coefficient"], color="skyblue")
plt.gca().invert_yaxis()  # Supaya ranking terbesar di atas
plt.xlabel("Absolute Coefficient")
plt.title("Ranking Pengaruh Variabel terhadap Crop Yield")
plt.tight_layout()
plt.show()
```

* **p-value = 0.4142**  Jauh di atas batas signifikansi umum ( = 0.05).

#### **Interpretasi**

    Artinya, hubungan antara emisi CO dan hasil panen sangat lemah dan cenderung negatif. Peningkatan emisi CO sedikit berkorelasi dengan penurunan hasil panen, namun kekuatan hubungan ini hampir tidak signifikan.

```python
from scipy.stats import pearsonr
import pandas as pd


corr_by_crop = pd.DataFrame(columns=['crop_type', 'r_CO2_vs_yield', 'r_CO2_vs_soil'])

for crop, sub in df.groupby('crop_type'):
    if len(sub) >= 3:
        r_yield, _ = pearsonr(sub['co2_emissions_mt'], sub['crop_yield_mt_per_ha'])
        r_soil, _  = pearsonr(sub['co2_emissions_mt'], sub['soil_health_index'])
        corr_by_crop.loc[len(corr_by_crop)] = [crop, r_yield, r_soil]

corr_by_crop = corr_by_crop.sort_values('r_CO2_vs_yield', ascending=False).reset_index(drop=True)

# Menambahkan kolom urutan
corr_by_crop.index = corr_by_crop.index + 1

print("Korelasi CO dengan Hasil Panen & Kesehatan Tanah per Jenis Tanaman:")
display(corr_by_crop)
```

1. Korelasi Negatif terhadap Hasil Panen
Hampir semua jenis tanaman menunjukkan korelasi negatif antara emisi CO dan produktivitas. Walaupun nilainya kecil (weak correlation), ini mengindikasikan bahwa peningkatan emisi CO kemungkinan memiliki efek menurunkan hasil panen, terutama pada Rice dan Coffee.

## Question 7: Tren produktivitas pertanian tiap benua dari 19902024 & proyeksinya hingga 2030

```python
# Mapping negara ke benua
continent_map = {
    'India': 'Asia',
    'China': 'Asia',
    'France': 'Europe',
    'Russia': 'Europe',
    'Canada': 'North America',
    'USA': 'North America',
    'Brazil': 'South America',
    'Argentina': 'South America',
    'Nigeria': 'Africa',
    'Australia': 'Oceania'
}

# Menambahkan kolom benua
df['Continent'] = df['Country'].map(continent_map)

# agregarsi data/benua/tahun
df_continent = df.groupby(['Year', 'Continent']).agg({
    'Crop_Yield_MT_per_HA': 'mean'
}).reset_index()
```

Analisis ini menggunakan metode ARIMA (AutoRegressive Integrated Moving Average) untuk memproyeksikan produktivitas pertanian (Crop Yield, dalam ton per hektar) berdasarkan data historis 19902024. Proyeksi dilakukan hingga tahun 2030 untuk setiap benua yang ada di dataset.

```python
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

forecast_results = {}

for continent in df_continent['Continent'].unique():
    temp = df_continent[df_continent['Continent'] == continent].set_index('Year')
    series = temp['Crop_Yield_MT_per_HA']

    # Fit ARIMA(1,1,1) - bisa di-tuning
    model = ARIMA(series, order=(1,1,1))
    model_fit = model.fit()

    # Forecast 6 tahun ke depan
    forecast = model_fit.forecast(steps=6)
    forecast_results[continent] = forecast

    # Plot
    plt.figure(figsize=(8,4))
    plt.plot(series, label='Actual')
    plt.plot(range(series.index.max()+1, series.index.max()+1+len(forecast)), forecast, label='Forecast', color='red')
    plt.title(f"Forecast Crop Yield - {continent}")
    plt.xlabel("Year")
    plt.ylabel("Crop Yield (MT per HA)")
    plt.legend()
    plt.show()
```

## **Rekomendasi**

**1. Fokus Adaptasi pada Wilayah & Komoditas Rawan**
- Prioritaskan Midwest, Patagonia, sayuran, dan kopi untuk intervensi adaptasi iklim.

**2. Investasi dalam Teknologi Pertanian**
- Gunakan irigasi presisi, varietas tahan iklim ekstrem, dan pemantauan iklim berbasis IoT.

**3. Pendekatan Multi-Faktor**
- Karena input tunggal seperti pupuk atau irigasi tidak berdampak besar secara independen, pendekatan terpadu (kombinasi teknologi, manajemen lahan, dan pemantauan cuaca) perlu diterapkan.

**4. Pengendalian Emisi CO**
- Kurangi emisi dari praktik pertanian intensif, sambil mengintegrasikan metode ramah lingkungan seperti agroforestri dan pertanian konservasi.

**5. Monitoring Produktivitas Jangka Panjang**
- Lakukan evaluasi berkala tren produktivitas untuk mengantisipasi penurunan dan mengukur keberhasilan strategi adaptasi.
